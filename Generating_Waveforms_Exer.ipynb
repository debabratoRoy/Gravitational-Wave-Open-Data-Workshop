```python
from pycbc.waveform import get_td_waveform
import matplotlib.pyplot as plt
```

## Exercise

* Generate and plot the waveform associated to the binary neutron star merger GW170817. Look up the GWTC-1 catalog page to obtain the estimated parameters for this source.

Here,
For GW170817,  
mass1 = 1.46 solar mass, mass2 = 1.27 solar mass,  
chirp mass = 1.186 solar mass,  
the analysis starts at f_lower = 23 Hz

```python

hp, hc = get_td_waveform(approximant="IMRPhenomD",
                         mass1=1.46,
                         mass2=1.27,
                         delta_t=1.0/16384,
                         f_lower=23)

plt.figure(figsize = (12, 4))
plt.plot(hp.sample_times,hp,label ='Plus Polarization')
plt.plot(hc.sample_times,hc,label ='Cross Polarization')
plt.xlabel('Time (s)')
plt.ylabel('Strain')
plt.legend()
plt.grid()
plt.show()

# Zoom near merger
plt.figure(figsize = (12, 4))
plt.plot(hp.sample_times,hp,label ='Plus Polarization')
plt.plot(hc.sample_times,hc,label ='Cross Polarization')
plt.xlim(-0.01,0.01)
plt.xlabel('Time (s)')
plt.ylabel('Strain')
plt.legend()
plt.grid()
plt.show()
```

* How much time does the signal last using a lower frequency of 30 Hz?

```python
# The masses are m1 = 1.46 solar mass, m2 = 1.27 solar mass and chirm mass, M = 1.186 solar mass

hp, hc = get_td_waveform(approximant="IMRPhenomD",
                         mass1=1.46,
                         mass2=1.27,
                         delta_t=1.0/16384,
                         f_lower=30)

plt.figure(figsize = (12, 4))
plt.plot(hp.sample_times,hp,label ='Plus Polarization')
plt.plot(hc.sample_times,hc,label ='Cross Polarization')
plt.xlabel('Time (s)')
plt.ylabel('Strain')
plt.legend()
plt.grid()
plt.show()

# Zoom near merger
plt.figure(figsize = (12, 4))
plt.plot(hp.sample_times,hp,label ='Plus Polarization')
plt.plot(hc.sample_times,hc,label ='Cross Polarization')
plt.xlim(-0.01,0.01)
plt.xlabel('Time (s)')
plt.ylabel('Strain')
plt.legend()
plt.grid()
plt.show()
```
